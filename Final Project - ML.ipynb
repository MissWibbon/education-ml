```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#read csv
SY19 = pd.read_csv("/Users/yxair/Desktop/NW Bootbcamps/HW/Group Project-3/CollegeScorecard_Raw_Data/MERGED2018_19_PP.csv")
# SY19_IL.head()
SY19DF= SY19[["INSTNM","ADM_RATE","UGDS","SAT_AVG","COSTT4_A","TUITIONFEE_OUT","AVGFACSAL","D_PCTPELL_PCTFLOAN","GRADS","RET_FT4_POOLED_SUPP","RET_PT4_POOLED_SUPP","OMAWDP8_FULLTIME","C100_4","SCHTYPE","HIGHDEG","CCSIZSET"]]

SY19DF = SY19DF.dropna()
SY19Cldt=SY19DF.rename(columns={"ADM_RATE":"Admission rate","UGDS":"Enrollment","SAT_AVG":"Average SAT","COSTT4_A":"Average cost","TUITIONFEE_OUT":"Out-of-state tuition ","AVGFACSAL":"Average faculty salary","D_PCTPELL_PCTFLOAN":"Number of undergraduate","GRADS":"Number of graduate","RET_FT4_POOLED_SUPP":"full-time student","RET_PT4_POOLED_SUPP":"part-time student","OMAWDP8_FULLTIME":"% receiving an award","C100_4":"Completion Rate","SCHTYPE":"Private","HIGHDEG":"Highest degree","CCSIZSET":"Size and Setting"})


SY19Cldt.head()

```

```python
SY19_KM=SY19_df.drop('Private',axis=1)
SY19_KM.head()
```

# K-Means Cluster for unsupervised learning

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(SY19_KM)
```

```python
predicted_clusters = kmeans.predict(SY19_KM)
```

```python
plt.scatter(SY19_KM.iloc[:, 2], SY19_KM.iloc[:, 3], c=predicted_clusters, s=50, cmap='viridis')
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(confusion_matrix(SY19_df['Private'],kmeans.labels_))
print(classification_report(SY19_df['Private'],kmeans.labels_))
```

# Logistic Regression for supervised learning

```python
#logistic regression for supervised learning
from sklearn.model_selection import train_test_split

X = SY19_df.drop('Private',axis=1)
y = SY19_df['Private']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
classifier.coef_
```

```python
feature_names=list(SY19_KM.columns)
sorted(zip(classifier.coef_[0], feature_names), reverse=True)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

# unsupervised and supervised model comparsion 

```python
print("Using K means clustering (unsupervised):\n")
print(confusion_matrix(SY19_df['Private'],kmeans.labels_))
print(classification_report(SY19_df['Private'],kmeans.labels_))

print("Using logistic regression (supervised):\n")
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```

# Create a Random Forest Classifier

```python
# Create a random forest classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
X = SY19_df.drop('Private',axis=1)
y = SY19_df['Private']
rf = rf.fit(X, y)
rf.score(X, y)
```

```python
rf_importances = rf.feature_importances_
rf_importances
```

```python
feature_names=list(SY19_KM.columns)
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
rf_sorted_idx = np.argsort(rf_importances)
pos = np.arange(rf_sorted_idx.shape[0]) + .5
fig = plt.figure(figsize=(12, 6))
plt.barh(pos, rf_importances[rf_sorted_idx], align='center')
plt.yticks(pos, np.array(feature_names)[rf_sorted_idx])
plt.title('Random Forest Classifier Feature Importance (MDI)')
```

# Create Gradient Boosting Regression

```python
from sklearn import ensemble
from sklearn.inspection import permutation_importance
from sklearn.metrics import mean_squared_error
```

```python
X = SY19_df.drop('Private',axis=1)
y = SY19_df['Private']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

params = {'n_estimators': 500,
          'max_depth': 4,
          'min_samples_split': 5,
          'learning_rate': 0.01,
          'loss': 'ls'}
```

```python
reg = ensemble.GradientBoostingRegressor()
reg
```

```python
reg.fit(X_train, y_train)
```

```python
mse = mean_squared_error(y_test, reg.predict(X_test))
print(f'The mean squared error (MSE) on test set: {mse}')
```

```python
gbs_feature_importance = reg.feature_importances_
```

```python
gbs_sorted_idx = np.argsort(gbs_feature_importance)
pos = np.arange(gbs_sorted_idx.shape[0]) + .5
fig = plt.figure(figsize=(12, 6))
plt.barh(pos, gbs_feature_importance[gbs_sorted_idx], align='center')
plt.yticks(pos, np.array(feature_names)[gbs_sorted_idx])
plt.title('Gradient Boosting Regression Feature Importance (MDI)')
```

```python
result = permutation_importance(reg, X_test, y_test, n_repeats=10,
                                random_state=42, n_jobs=2)
sorted_idx = result.importances_mean.argsort()
plt.boxplot(result.importances[sorted_idx].T,
            vert=False, labels=np.array(feature_names)[sorted_idx])
plt.title("Permutation Importance (test set)")
fig.tight_layout()
plt.show()
```
